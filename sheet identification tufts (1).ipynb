```python
import pandas as pd
import os
from openpyxl import load_workbook

# Load Excel file
file_path = r"C:\Users\shres\Downloads\TUFTS 3 typess.xlsx"
xls = pd.ExcelFile(file_path)

# Create directories to store the output files
output_dir_tables = r"C:\Users\shres\Downloads\Extracted_Tables_Tuftssser"
output_dir_freedom = r"C:\Users\shres\Downloads\Freedom_Worksheetss"
output_dir_attributes = r"C:\Users\shres\Downloads\Attributes_Worksheetss"

if not os.path.exists(output_dir_tables):
    os.makedirs(output_dir_tables)

if not os.path.exists(output_dir_freedom):
    os.makedirs(output_dir_freedom)

if not os.path.exists(output_dir_attributes):
    os.makedirs(output_dir_attributes)

# Define the strings with leading and trailing spaces for table extraction
start_string = 'UNITED HEALTHCARE UNDERWRITING FULLY INSURED APPROVAL FORM - For Internal Use Only'
end_string = 'GRAND TOTAL'

# Iterate through each sheet in the Excel file
for sheet_name in xls.sheet_names:
    print(f"Processing sheet: {sheet_name}")
    
    # Read the sheet into a DataFrame
    df = pd.read_excel(file_path, sheet_name=sheet_name, header=None)
    
    # Check if the value in the first row and first column contains the substring "Freedom"
    if df.iloc[0, 0].find("Freedom") != -1:
        # Save the worksheet into a new Excel file in Freedom worksheets directory
        output_file_path_freedom = os.path.join(output_dir_freedom, f"{sheet_name}.xlsx")
        df.to_excel(output_file_path_freedom, index=False, header=False)
        print(f"Worksheet starting with 'Freedom' saved to: {output_file_path_freedom}")
    elif df.iloc[0, 0].find("Attributes") != -1:
        # Save the worksheet into a new Excel file in Attributes worksheets directory
        output_file_path_attributes = os.path.join(output_dir_attributes, f"{sheet_name}.xlsx")
        df.to_excel(output_file_path_attributes, index=False, header=False)
        print(f"Worksheet starting with 'Attributes' saved to: {output_file_path_attributes}")
    else:
        # Initialize flags to identify the start and end of the table for non-Freedom worksheets
        table_started = False
        
        # Iterate through rows in the DataFrame
        for index, row in df.iterrows():
            # Check if the start string is found in the row
            if start_string in ' '.join(map(str, row)):
                start_row_index = index
                table_started = True
            # Check if the end string is found in the row and the table has started
            elif table_started and end_string in ' '.join(map(str, row)):
                end_row_index = index
                # Select the table rows between start and end row indexes
                selected_table = df.iloc[start_row_index:end_row_index + 1]
                
                # Save the selected table into a new Excel file with the same sheet name
                output_file_path_tables = os.path.join(output_dir_tables, f"{sheet_name}.xlsx")
                selected_table.to_excel(output_file_path_tables, index=False, header=False)
                print(f"Extracted table saved to: {output_file_path_tables}")
                
                # Rename the sheet with the sheet name
                wb = load_workbook(output_file_path_tables)
                ws = wb.active
                ws.title = sheet_name
                wb.save(output_file_path_tables)
                
                # Reset the flag
                table_started = False
                # Break the loop as we found the end of the table
                break
```
